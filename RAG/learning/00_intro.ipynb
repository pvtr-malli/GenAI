
# **RAG (Retrieval-Augmented Generation) – Simple Steps**

1. Collect and clean documents
2. Split text into chunks
3. Create embeddings for each chunk
4. Store embeddings in a vector database
5. User asks a question
6. Convert the question into an embedding
7. Retrieve most relevant chunks
8. Combine retrieved text with the question
9. Send to LLM for answer generation
10. Return grounded, context-aware response
